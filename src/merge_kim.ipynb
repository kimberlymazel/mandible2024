Copy Files (with class replacement)

```python
import os
import shutil

source_folder = "../dataset/roboflow/noneed/valid/labels"
destination_folder = "../dataset/roboflow/noneed_filtered/labels"

for filename in os.listdir(source_folder):
    if filename.endswith(".txt"):  
        source_file_path = os.path.join(source_folder, filename)
        
        with open(source_file_path, "r") as file:
            content = file.readlines()
        
        filtered_content = [line for line in content if line.split()[0] == "0"]
        
        destination_file_path = os.path.join(destination_folder, filename)
        
        with open(destination_file_path, "w") as file:
            file.writelines(filtered_content)
        
        print(f"Filtered dataset created successfully for {filename}.")

print("All files processed.")
```

```python
import os

source_folder = "../dataset/roboflow/yolonihadatsetobject/valid/labels"
destination_folder = "../dataset/roboflow/yolonihadatsetobject_filtered/labels"

for filename in os.listdir(source_folder):
    if filename.endswith(".txt"):
        source_file_path = os.path.join(source_folder, filename)
        
        with open(source_file_path, "r") as file:
            content = file.readlines()

        filtered_content = [line.replace("5", "0", 1) for line in content if line.split()[0] == "5"]
        
        destination_file_path = os.path.join(destination_folder, filename)
        
        with open(destination_file_path, "w") as file:
            file.writelines(filtered_content)
        
        print(f"Filtered dataset created successfully for {filename}.")

print("All files processed.")
```

```python
import os
import shutil

source_folder = "../dataset/roboflow/lko/train/labels"
destination_folder = "../dataset/roboflow/lko_filtered/labels"

for filename in os.listdir(source_folder):
    if filename.endswith(".txt"): 
        source_file_path = os.path.join(source_folder, filename)
        
        with open(source_file_path, "r") as file:
            content = file.readlines()
        
        modified_content = []
        for line in content:
            class_label, *rest = line.split()
            if class_label == "0":
                modified_content.append(line)
            elif class_label == "1":
                modified_content.append("0 " + " ".join(rest) + "\n")
        
        destination_file_path = os.path.join(destination_folder, filename)

        with open(destination_file_path, "w") as file:
            file.writelines(modified_content)
        
        print(f"Filtered dataset created successfully for {filename}.")

print("All files processed.")
```

## Count File No. in Directory

```python
import os

directory_path = "../dataset/annotations"

file_count = 0

for filename in os.listdir(directory_path):
    if os.path.isfile(os.path.join(directory_path, filename)):
        file_count += 1

print("Total number of files in the directory:", file_count)
```

## Compare Images and Labels

```python
import os
import pandas as pd

directory1_path = "../dataset/merged/merged2/images"
directory2_path = "../dataset/merged/merged2/labels"

directory1_files = sorted(os.listdir(directory1_path))
directory2_files = sorted(os.listdir(directory2_path))

max_length = max(len(directory1_files), len(directory2_files))

df = pd.DataFrame({'Directory 1': directory1_files + [''] * (max_length - len(directory1_files)),
                   'Directory 2': directory2_files + [''] * (max_length - len(directory2_files))})

csv_file_path = "comparison_table2.csv"
df.to_csv(csv_file_path, index=False)

print(f"DataFrame exported to CSV file: {csv_file_path}")
```
