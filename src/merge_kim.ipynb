```python
import os
import shutil

# Specify the folder containing the class files
source_folder = "../dataset/roboflow/valid/labels"
destination_folder = "../dataset/roboflow_filtered/labels"

# Iterate through all files in the folder
for filename in os.listdir(source_folder):
    if filename.endswith(".txt"):  # Process only text files
        source_file_path = os.path.join(source_folder, filename)
        
        # Read the content of the current file
        with open(source_file_path, "r") as file:
            content = file.readlines()
        
        # Filter out the entries with class label 1 (maxilla) from the two classes dataset
        filtered_content = [line for line in content if line.split()[0] == "0"]
        
        # Create the destination file path
        destination_file_path = os.path.join(destination_folder, filename)
        
        # Write the filtered content to a new file in the destination folder
        with open(destination_file_path, "w") as file:
            file.writelines(filtered_content)
        
        print(f"Filtered dataset created successfully for {filename}.")
        # Optionally, you can delete the original file
        os.remove(source_file_path)

print("All files processed.")
```

```python
import os
import shutil

# Specify the source and destination folders
source_folder = "../xray_panoramic_mandible/images"
destination_folder = "../dataset/merged/images"

# Iterate through all files in the source folder
for filename in os.listdir(source_folder):
    if filename.endswith(".png"):  # Process only text files
        source_file_path = os.path.join(source_folder, filename)
        destination_file_path = os.path.join(destination_folder, filename)
        
        # Copy the file to the destination folder
        shutil.copy(source_file_path, destination_file_path)
        
        print(f"File {filename} copied successfully to {destination_folder}.")

print("All files copied.")
```
